```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
import seaborn as sns
```

## Tarea 1



```python
u = np.random.uniform(size=500)
```

```python
def f(vector):
  return 1/(1-vector)**(0.25)
```

```python
x = f(u)
```

```python
1/(1-u[0])**0.25 == x[0]
```

```python
a,m = 4, 1
fig, ax = plt.subplots(figsize=(14,7))
count, bins, _  = plt.hist(x,bins=500, density=True, label = 'Simulacin')
plt.plot(bins, a*m**a/(bins**(a+1)), color = 'red', label = 'pdf Pareto')
plt.legend()
plt.title('Histograma de Densidad')
```

```python
# Medidas de la simulacion
mu_sim = x.mean()
var_sim = x.var()
print('media simulacin : ', mu_sim)
print('varianza simulacin : ', var_sim)
```

```python
mu_pareto = m*1**(1/a)
var_pareto = 1*a/(((a-1)**2)*(a-2))
print('media pareto : ', mu_pareto)
print('varianza pareto : ', var_pareto)
```

## Tarea 2


1. Tome  = 2. Simule 10 mil muestras de T1 y T2, para cada par (T1, T2) obtenga (C, U). De este modo tendr 10 mil muestras de (C, U).

```python
lambda_ = 2
T1 = np.random.exponential(1/lambda_, 10000)
T2 = np.random.exponential(1/lambda_, 10000)


C = T1 + T2
U = T1/(T1+T2)

```

```python
len(C), len(U)
```

2. Haga el histograma de densidad para las muestras obtenidas de C, y superponga la densidad de Gamma. Haga lo mismo para U con la densidad uniforme.

```python
fig, ax = plt.subplots(figsize=(14,7))
count, bins, _  = plt.hist(C,bins = 100, density=True, label = 'Simulacin')
plt.plot(bins,  ss.gamma.pdf(bins, 2, scale = 1/lambda_) , color = 'red', label = 'pdf')
plt.legend()
plt.title('Histograma de Densidad')
```

```python
fig, ax = plt.subplots(figsize=(14,7))
count, bins, _  = plt.hist(U,bins = 100, density=True, label = 'Simulacin')
plt.plot(bins,  ss.uniform.pdf(bins), color = 'red', label = 'pdf')
plt.legend()
plt.title('Histograma de Densidad')
```

3.  Calcule la correlacin entre C y U. Entregue un grfico con los pares (c1, u1), . . . , (cs, us) donde s = 10,000.

```python
fig, ax = plt.subplots(figsize=(14,7))
plt.scatter(C,U)
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(C, U)
corr
```
